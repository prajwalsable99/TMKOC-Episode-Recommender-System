```python
data =[]
```

```python

for ind in df.index:
    # print(df['Episode_title'][ind], df['description'][ind])
    data.append( str(df['Episode_title'][ind]) + str( df['description'][ind]))
    
```

```python
data
```

```python
len(data)
newdata=[]
for row in data :
    newdata.append(re.sub("\d+", " ", row))
data=newdata
```

```python
vocabulary=set()
for sentence in data:
    words= sentence.split(' ')
    for word in words: 
        vocabulary.add(word)
len(vocabulary)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create object
tfidf = TfidfVectorizer(stop_words='english')

# get tf-df values
sparse_matrix = tfidf.fit_transform(data)
```

```python
tfidf_matrix= sparse_matrix.toarray()
```

```python
tfidf_df= pd.DataFrame(tfidf_matrix, columns= tfidf.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi=cosine_similarity(sparse_matrix)
```

```python
simi.shape
```

```python
final_matrix= pd.DataFrame(simi,index=df['Episode_title'],columns=df['Episode_title'])
```

```python
final_matrix
```

```python

def get_top_recommendations(input_text, vectorizer, sparse_matrix, top_n=10):
    input_tfidf = vectorizer.transform([input_text])
    cosine_similarities = cosine_similarity(input_tfidf, sparse_matrix).flatten()
    top_indices = cosine_similarities.argsort()[-top_n:][::-1]
    recs=[]   
    
    for index in top_indices:
         recs.append(   df.iloc[index])

    recs=pd.DataFrame(recs).reset_index(drop=True)
    
    
    return recs
```

```python
# Example usage
input_text = "ganesh"
top_recommendations = get_top_recommendations(input_text, tfidf, sparse_matrix)
```
